```python
import os
import glob
import os
import large_image as li
import cv2
import argparse
import pandas as pd
json_path = '/home/bnc/tool/HistomicsML/yourproject/keras-yolo3/wsi_json/'
wsi_path = '/home/bnc/tool/HistomicsML/yourproject/WSI/'
save_path = '/home/bnc/tool/HistomicsML/yourproject/keras-yolo3/image/'
json_list = os.listdir(json_path)
wsi_list = os.listdir(wsi_path)
print(len(json_list))
print(len(wsi_list))
```

```python
def resize(wsi_path,save_path,magn=None):
    ts = li.getTileSource(wsi_path)
    info = ts.getMetadata()
    filename = os.path.basename(wsi_path)
    file = filename.split('.')[0][5:]
    sz = (info['sizeX'],info['sizeY'])
    im_low_res, _ = ts.getRegion(
            scale = dict(magnification=magn),
            format = li.tilesource.TILE_FORMAT_NUMPY)
    cv2.imwrite(save_path+"{}.png".format(file),im_low_res)
    print("{} has been resize".format(file))
```

```python
def main():
    cnt = 0
    for json in json_list:
        json_name = json.split('.')[0]
        for wsi in wsi_list:
            wsi_name = wsi.split('.')[0][5:]
            if json_name == wsi_name:
                resize(os.path.join(wsi_path,wsi),save_path,0.3125)
                cnt = cnt + 1
                print(cnt)
```

```python
if __name__ == "__main__":
    main()
```
